```python
class Patient:
    def __init__(self, name, age, gender, diagnosis):
        self.name = name
        self.age = age
        self.gender = gender
        self.diagnosis = diagnosis

    def __str__(self):
        return f"ID: {self.id} | Name: {self.name} | Age: {self.age} | Gender: {self.gender} | Diagnosis: {self.diagnosis}"

class Hospital:
    def __init__(self):
        self.patients = {}
        self.next_id = 1

    def add_patient(self, name, age, gender, diagnosis):
        patient = Patient(name, age, gender, diagnosis)
        patient.id = self.next_id
        self.patients[self.next_id] = patient
        self.next_id += 1
        print(f"Patient {patient.id} added successfully!")

    def view_all_patients(self):
        if not self.patients:
            print("No patients found.")
        for patient in self.patients.values():
            print(patient)

    def view_patient(self, patient_id):
        patient = self.patients.get(patient_id)
        if patient:
            print(patient)
        else:
            print(f"Patient with ID {patient_id} not found.")

    def update_patient(self, patient_id, name=None, age=None, gender=None, diagnosis=None):
        patient = self.patients.get(patient_id)
        if patient:
            if name:
                patient.name = name
            if age:
                patient.age = age
            if gender:
                patient.gender = gender
            if diagnosis:
                patient.diagnosis = diagnosis
            print(f"Patient {patient_id} updated successfully!")
        else:
            print(f"Patient with ID {patient_id} not found.")

    def delete_patient(self, patient_id):
        if patient_id in self.patients:
            del self.patients[patient_id]
            print(f"Patient {patient_id} deleted successfully!")
        else:
            print(f"Patient with ID {patient_id} not found.")

    
hospital = Hospital()
hospital.add_patient("Krish", 35, "Male", "Flu")
hospital.add_patient("Kaira", 28, "Female", "Fracture")
hospital.add_patient("Sayli",23,"Female","Cold")
hospital.view_all_patients()
hospital.update_patient(1, diagnosis="Recovered from Flu")
hospital.view_patient(1)
hospital.delete_patient(2)
hospital.view_all_patients()
```
